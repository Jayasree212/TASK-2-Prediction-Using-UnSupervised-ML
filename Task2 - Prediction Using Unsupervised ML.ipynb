**Step 4: Applying K-Means Clustering algorithm**

```python
#Extracting the values of different attributes
x=iris_df.iloc[:,[0,1,2,3]].values
x
```

```python
from sklearn.cluster import KMeans
#Sum of squares with in clusters
WCSS=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)
```

**Step 5: Plotting K-Means Clustering Graph**

```python
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

In Elbow Method, the elbow occurs where the optimum number of clusters are present. </br>
From the above graph using the"Elbow method", we conclude that the optimum number of clusters are to be 3. 


```python
kmeans=KMeans(n_clusters=3,init='k-means++', max_iter=300, n_init=10,  random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

**Step 6: Visualizing the clusters**

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1],s=80, c="red", label ="Iris-Setosa")

plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1],s=80, c="blue", label ="Iris-Versicolour")

plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1],s=80, c="green", label ="Iris-Virginica")

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c="yellow", label="centroids")
plt.legend()
```

**Step 7: Make Predictions**

**Step 8: Evaluate the model**

```python
# Comaparing actual vs predicted values
new_labels=kmeans.labels_
new_labels
```

```python
#Plotting the identified clusters and comparing with the results

fig,axes=plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(x[:,0],x[:,1], c=y, cmap="gist_rainbow", edgecolor="k", s=80)
axes[1].scatter(x[:,0],x[:,1], c=new_labels, cmap="viridis", edgecolor="k", s=80)

axes[0].set_xlabel('Sepal Length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)

axes[1].set_xlabel('Sepal Length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)

axes[0].tick_params(direction="in",length=10, width=5, colors="k", labelsize=20)
axes[1].tick_params(direction="in",length=10, width=5, colors="k", labelsize=20)

axes[0].set_title("Actual", fontsize=18)
axes[1].set_title("Predicted", fontsize=18)

plt.show()

```

From the above graphs we can conclude that the Actual and The Predicted values are mostly same.
